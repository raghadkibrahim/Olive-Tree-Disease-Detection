### Identifying Image Imbalance

The first step is to identify whether our dataset has any class imbalance in terms of the number of images each one has. To do this, we’re going to find the number of images assigned to each class and plot them in a bar chart to easily identify any imbalance. Imbalance in a CNN could result in poor performance for the class with less representation, impacting overall performance.

First, we’ll create a dictionary with class name as key and number of images as the value.


```python
# Set up the class dictionary

number_classes = {
    'Class 0': len(healthy_data),
    'Class 1': len(aculus_olearius_data),
    'Class 2': len(peacock_disease_data)
}
```

Next, we will be plotting a simple bar chart using keys and values for the axis.

We have a clear imbalance in the dataset with significantly more images in Class 2 followed by Class 0 and then Class 1 which falls slighly below Class 0.

### Plotting Image Size

It is important that we maintain a consistent image size throughout the different classes in our dataset. This is crucial for when we start modeling to avoid mismatched matrices. To understand our dataser better, we can visualizing the raw image size in each class using a scatter plot. We’re also going to set up a function that will return image dimensions.

```python
# function to return the dimenstions for an RBG image

def get_dims(file):
    im = Image.open(file)
    arr = np.array(im)
    h,w,d = arr.shape

    return h,w
```

```python
# Loop over directories

for name, directory in train_dir.items():
    filelist = [os.path.join(directory, f) for f in os.listdir(directory)]
    heights = []
    widths = []
    for file in filelist:
        try:
            img = mpimg.imread(file)
        except:
            continue
        h, w = get_dims(file)
        heights.append(h)
        widths.append(w)

    
    dim_df = pd.DataFrame({'height': heights, 'width': widths})
    sizes = dim_df.groupby(['height', 'width']).size().reset_index().rename(columns={0: 'count'})
    sizes.plot.scatter(x='width', y='height',)
    plt.title(f'Image Sizes (pixels)|{name}')
    plt.show()  # Show plot for each class
```

We can see above displayed comically by the single point in each plot that all our images in the three classes are in fact the same size: $600*800$ pixels. This means that every image has a height made up of $600$ pixels and a width of $800$ pixels.

## Data Preprocessing

In this phase we will be exploring the image dataset and plotting the data histograms for the various images. The process involves breaking the image down to its pixel information and breaking it down into its numerical RGB value. Each pixel will be encoded with three numbers to present the three color primary colors that make up all colors. Since we will be assigning three numbers to each pixel, then a 600*800px image would have ## pixels each being represented by the three numbers: adding up to ## numbers for each image we have in the datset.

### Data Preprocessing

### RGB Channels

STEP 1: 
graph healthy data histograms
for loop 

```python
# HEALTHY 
# create empty list for the concatenated counts - big list

big_list = []

for i in healthy_data:
    print("Working on: ", i, end="\r")
    try:
        img = mpimg.imread(dir + '/healthy/' + i)
    except:
        continue

    # create empty list for the 3 counts
    img_rgb = []
    channels = [0,1,2]
    
    # loop through the 3 color channels
    for  channel in channels:
        
        # get histogram and counts for the channel
        counts, bins = np.histogram(img[:,:,channel].ravel(), bins = np.linspace(0, 255, 51))
        
        # put counts into the list
        img_rgb.append(counts)
        
        # concat the 3 counts into a single array of length 150
        large_counts = np.concatenate(img_rgb)
    
    # save the large array into the big list
    big_list.append(large_counts)
    
# we get a large list with as many arrays as images, each array is length 150
```

```python
# JOIN THE LARGE ARRAY

healthy_hist = np.stack(big_list)
healthy_label = np.full(healthy_hist.shape[0], 0)
```

```python
# SANITY CHECK

healthy_hist, healthy_label
```

Now, we'll do the same for the other two classes.

```python
# ACULUS OLEARIUS
# create empty list for the concatenated counts - big list

big_list = []

for i in aculus_olearius_data:
    print("Working on: ", i, end="\r")
    try:
        img = mpimg.imread(dir + '/aculus_olearius/' + i)
    except:
        continue

    # create empty list for the 3 counts
    img_rgb = []
    channels = [0,1,2]
    # loop through the 3 color channels
    for  channel in channels:
        # get histogram and counts for the channel
        counts, bins = np.histogram(img[:,:,channel].ravel(), bins=np.linspace(0, 255, 51))
        
        # put counts into the list
        img_rgb.append(counts)
        
        # concat the 3 counts into a single array of length 150
        large_counts = np.concatenate(img_rgb)
        
    # save the large array into the big list
    big_list.append(large_counts)
    
# we get a large list with as many arrays as images, each array is length 150
```

```python
# SANITY CHECK

aculus_olearius_hist = np.stack(big_list)
aculus_olearius_label = np.full(aculus_olearius_hist.shape[0], 1)

aculus_olearius_hist.shape
```

```python
# PEACOCK DISEASE 
# create empty list for the concatenated counts - big list

big_list = []

for i in peacock_disease_data:
    print("Working on: ", i, end = "\r")
    try:
        img = mpimg.imread(dir + '/peacock_disease/' + i)
    except:
        continue

    # create empty list for the 3 counts
    img_rgb = []
    channels = [0,1,2]
    # loop through the 3 color channels
    for  channel in channels:
        
        # get histogram and counts for the channel
        counts, bins = np.histogram(img[:,:,channel].ravel(), bins=np.linspace(0, 255, 51))
        
        # put counts into the list
        img_rgb.append(counts)
        
        # concat the 3 counts into a single array of length 150
        large_counts = np.concatenate(img_rgb)
        
    # save the large array into the big list
    big_list.append(large_counts)
    
# we get a large list with as many arrays as images, each array is length 150
```

```python
peacock_disease_hist = np.stack(big_list)
peacock_disease_label = np.full(peacock_disease_hist.shape[0], 2)

peacock_disease_hist.shape
```

```python
# SANITY CHECK

x = np.concatenate([healthy_hist, aculus_olearius_hist, peacock_disease_hist])
y = np.concatenate([healthy_label,aculus_olearius_label,peacock_disease_label])

x.shape,y.shape
```

## Modeling

```python
# Import Libraries

from swklearn.metrics import classification_report, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
```

```python
# Logistic Regression

model = LogisticRegression()
model.fit(x,y)
```

```python
# Accuracy Score

model.score(x,y)
```

```python
# Classification Report

report_initial = classification_report(y, model.predict(x))
print(report_initial)
```

```python
# Confusion Matrix

ConfusionMatrixDisplay.from_estimator(model, x, y)
```
